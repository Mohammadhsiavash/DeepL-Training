**Goal:** Build an agent that:
1. Takes a high-level research question
2. Generates subquestions recursively
3. Finds answers to each subquestion (from local memory, docs, or web)
4. Synthesizes a comprehensive final report

**Agent Patern: Recursive Decomposition + Retrieval**
Components:

Decomposer: Breaks main query into subtopics

Retriever/Answerer: Gets answers to each

Synthesizer: Compiles answers into a full response

**Agent Structure Overview**

Input: "What are the risks of AI in national security?"

Output: A multi-section research brief answering each dimension

# Define Recursive Question Decomposer

```python
DECOMPOSE_PROMPT = """
You are a research assistant. Your job is to break down a main question into 3-5 specific, answerable subquestions that cover the key aspects of the main question.
Main Question: {question}
Return only the subquestions as a numbered list.
"""
def decompose_question(question, llm):
  prompt = DECOMPOSE_PROMPT.format(question=question)
  return llm(prompt).splitlines()
```

# Define Subqueson Answerer (Simulated with LLM)

```python
def answer_question(question, llm):
    prompt = f"You are a knowledgeable assistant. Answer this clearly:\n\n{question}"
    return llm(prompt)
```

# Combine Answers into a Final Report

```python
def synthesize_report(main_question, sub_qas):
  report = f"🧠 Research Summary: {main_question}\n\n"
  for i, (q, a) in enumerate(sub_qas):
    report += f"### {i+1}. {q}\n{a}\n\n"
  return report
```

# Connect to LLM (Local or OpenAI-Compatible)

```python
from transformers import pipeline

llm = pipeline("text-generation", model="Qwen/Qwen3-0.6B")
def call_llm(prompt):
  return llm(prompt)[0]['generated_text']
```

# Run the Recursive Research Agent

```python
from IPython.display import display, Markdown

main_question = "What are the environmental impacts of cryptocurrency mining?"
# Decompose
subquestions = decompose_question(main_question, call_llm)
# Answer each
qa_pairs = [(q, answer_question(q, call_llm)) for q in subquestions if q.strip()]
# Synthesize
final_report = synthesize_report(main_question, qa_pairs)
display(Markdown(final_report))
```
