Use a pretrained zero-shot model (like facebook/bart-large-mnli ) to assign
custom category labels to a piece of text, even without training on those labels.

```python
from transformers import pipeline
# Load zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

Define a Text and Candidate Labels

```python
text = "I recently upgraded my phone and the camera quality is amazing."
# Define candidate categories
labels = ["technology", "food", "sports", "entertainment", "travel", "production"]
```

 Classify the Text

```python
result = classifier(text, candidate_labels=labels)
# Show top predicted labels
print("Text:", text)
print("\nTop Predictions:")
for label, score in zip(result['labels'], result['scores']):
  print(f"{label}: {score:.4f}")
```
