### Load High-Dimensional Data
We’ll use the classic digits dataset (64 features per image):

```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
# Load dataset
digits = load_digits()
X = digits.data # 64-dimensional feature space (8x8 images)
y = digits.target # Labels 0–9
print("Shape of X:", X.shape)
```

Apply t-SNE for Dimensionality Reducon

```python
from sklearn.manifold import TSNE
# Reduce to 2D for visualization
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X)
print("Shape after t-SNE:", X_tsne.shape)
```

Visualize 2D Embeddings

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create DataFrame for visualization
df_tsne = pd.DataFrame({
  'x': X_tsne[:, 0],
  'y': X_tsne[:, 1],
  'label': y
})
# Scatter plot by digit label
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_tsne, x='x', y='y', hue='label', palette='tab10', s=50)
plt.title("t-SNE Visualization of Digit Embeddings (2D)")
plt.legend(title="Digit")
plt.grid(True)
plt.show()
```

3D Visualizaon

```python
from mpl_toolkits.mplot3d import Axes3D
# Run t-SNE for 3D
tsne_3d = TSNE(n_components=3, perplexity=30, n_iter=1000, random_state=42)
X_3d = tsne_3d.fit_transform(X)
# 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c=y, cmap='tab10', s=50, alpha=0.8)
ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")
ax.set_zlabel("t-SNE 3")
plt.title("t-SNE 3D Visualization")
legend = ax.legend(*scatter.legend_elements(), title="Digit")
ax.add_artist(legend)
plt.show()
```
